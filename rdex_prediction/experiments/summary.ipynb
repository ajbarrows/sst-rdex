```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import neurotools.plotting as ntp

from abcd_tools.utils.ConfigLoader import load_yaml
params = load_yaml('../parameters.yaml')
```

```python
df = (
    pd.read_csv(params['model_results_path'] + 'ridge_models_summary.csv')
    .drop_duplicates(subset='target')
)
(
    df[['target', 'mean_scores_r2', 'std_scores_r2']]
    .set_index('target')
    .multiply(100)
    .sort_values('mean_scores_r2', ascending=False)
)
```
